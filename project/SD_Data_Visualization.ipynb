```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/final_df.csv')
```

```python
df.sort_values('roi', ascending=False)
```

```python
df.sort_values('production_budget_x', ascending=False)
```

```python
sns.set(font_scale=2)
ax = sns.lmplot(x='production_budget_x', y='worldwide_gross', data=df, hue='studio_size', markers=["1", "x", "o"],
           height=10, aspect=1.75, legend=False);
plt.xlim(0, None)
plt.ylim(0, None)
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.legend(loc='upper left', title_fontsize='24', title='Studio Size');
```

Let's zoom in.

```python
sns.set(font_scale=2)
sns.lmplot(x='production_budget_x', y='worldwide_gross', data=df, hue='studio_size', markers=["1", "x", "o"],
           height=10, aspect=1.75, legend=False);
plt.xlim(0, 10**8)
plt.ylim(0, 5*10**8)
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.legend(loc='upper left', title_fontsize='24', title='Studio Size');
```

Obvious rounding artificats.

Production budget of each movie vs. the worldwide gross.

```python
sns.set(font_scale=1)
sns.scatterplot(x='production_budget_x', y='roi', data=df, hue='studio_size')
           #markers=["1", "x", "o"],
           #height=10, aspect=1.75);
plt.xlim(0, None)
plt.ylim(-25, None)
plt.xlabel('Production Budget')
plt.ylabel('ROI')
plt.legend(loc='upper right', title='Studio Size');
```

Take a closer look.

```python
sns.set(font_scale=2)
sns.lmplot(x='production_budget_x', y='roi', data=df, hue='studio_size', markers=["1", "x", "o"],
           height=10, aspect=1.75, legend=False);
plt.xlim(0, 10**8)
plt.ylim(-5, 110)
plt.xlabel('Production Budget')
plt.ylabel('ROI')
plt.legend(loc='upper right', title_fontsize='24', title='Studio Size');
```

```python
plt.figure(figsize=(16, 10))
sns.regplot(x=df['production_budget_x'], y=df['roi'], fit_reg=False);
```

Larger studios have a larger mean ROI

```python
sns.set(font_scale=2)
plt.figure(figsize=(16, 8))
sns.boxplot(y=df['studio_size'], x=df['production_budget_x'])
plt.xlabel('Production Budget')
plt.ylabel('Studio Size');
```

Larger movie studios spent more money making movies and make more of them.

```python
sns.set(font_scale=2)
plt.figure(figsize=(16, 8))
sns.boxplot(y=df['studio_size'], x=df['roi'])
plt.xlabel('ROI')
plt.ylabel('Studio Size');
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(8, 4))
sns.boxplot(y=df['studio_size'], x=df['roi'])
plt.xlim(-2, 104)
plt.xlabel('ROI')
plt.ylabel('Studio Size');
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(10, 5))
sns.boxplot(y=df['studio_size'], x=df['roi'])
plt.xlim(-2, 65)
plt.xlabel('ROI')
plt.ylabel('Studio Size');
```

Not pictured: the ROI value near 400.

```python
sns.set(font_scale=2)
plt.figure(figsize=(16, 8))
sns.boxplot(y=df['studio_size'], x=df['worldwide_gross'])
plt.xlabel('Worldwide Gross')
plt.ylabel('Studio Size');
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(8, 4))
sns.barplot(x=df['movie_budget_size'], y=df['roi'], palette='husl', ci=None)
plt.xlabel('Movie Budget')
plt.ylabel('ROI');
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(8, 4))
sns.barplot(x='movie_budget_size', y='roi', data=df, hue='studio_size',ci=None)
plt.xlabel('Movie Budget')
plt.ylabel('ROI')
plt.legend(loc='upper left', title='Studio Size');
```
